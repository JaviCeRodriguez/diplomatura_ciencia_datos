## Ejercicio 3:


*El índice de masa corporal (IMC) se puede calcular utilizando la siguiente fórmula:*

> *IMC = P / A^2*

> *donde P es el peso de la persona en kilogramos y A es la altura de la misma persona en metros. La fórmula resulta en un número que se utiliza para determinar si una persona está en un peso saludable o no.* 

*Un IMC normal se encuentra entre 18.5 y 24.9 kg/m^2. Un IMC menor a 18.5 kg/m^2 se considera bajo peso, mientras que un IMC mayor a 25 kg/m^2 se considera sobrepeso. Un IMC mayor a 30 kg/m^2 se considera obesidad. Sin embargo, estos valores son solo una guía y no son válidos para todas las personas, especialmente aquellas con una gran cantidad de músculo o deportistas.*

**Crea un programa que te pida (usando input()) las siguientes informaciones: nombre, edad, peso en kg y altura en m y que te devuelva el nombre, la edad y el índice de masa corporal de esa persona.**


*PISTA: input() siempre devuelve siempre una cadena*

```python
# input() promptea al usuario y guarda esa variable COMO UNA CADENA (str).
nombre = input('¿Nombre?: ')

edad_str = input('¿Edad?: ')
edad = int(edad_str)  # Como input guarda la variable como cadena, hay que 'castearla' (convertir su tipo) a entero, con la funcion int() (int=integer=entero)
                      # ¿Que pasa si el usuario introduce una palabra en vez de un numero?

peso_str = input('¿Peso (en kg)?: ') 
peso = float(peso_str) # Como input guarda la variable como cadena, hay que 'castearla' (convertir su tipo) a flotante, con la funcion float() (float=flotante)
                       # OJO! float() solo entiende cuando los decimales se separan con '.' no con ','....
                       # ¿Habrá una manera de arreglar la cadena si el usuario usó ',' (antes de pasarla por float())? ¿Y de forzar al usuario a que introduzca si o si usando '.'?

altura_str = input('¿Altura (en m)?: ')
altura = float(altura_str)

masa_corporal = peso / (altura**2)

print(f'\nNombre: {nombre}\nEdad: {altura}\nIMC: {masa_corporal}')
```

## Ejercicio 5:

Crea un código que tome como parámetro una frase y sea capaz de contar el número de palabras que contiene. Haga un test con la frase: "Me gusta estudiar python en la UNSAM". 

Y si queremos contar los caracteres incluyendo los espacios. ¿Cómo seria?

Prueba con otras frases, para eso podes usar la funcion de ingreso de valores por teclado. 

```python
print('CONTADOR DE PALABRAS\n--------------------\n')

frase = input('Introduzca una frase:\n')
print() # print() siempre agrega al final de la impresion \n entonces pasa de renglon. Si uso print() sola, simplemente saltea el renglon.

frase_como_lista = frase.split(" ") # frase.split() va retornar una lista con la frase separada en cada espacio que encuentre, o sea que cada elemento de la lista es una palabra.
palabras = len(frase_como_lista) # Si yo veo el largo de una lista que tiene todas las palabras de la frase.... ENTONCES SE CUANTAS PALABRAS TIENE LA FRASE!

# También se puede resolver todo en un reglón! Es más elegante! En python, se pueden meter funciones adentro de funciones (composicion de funciones)
# Esta bueno hacerlo, siempre y cuando no se vuelva más dificil de leer. Sino es mejor hacerlo utilizando variables intermedias.
# palabras = len(frase.split(' '))

print(f'La frase tiene {palabras} palabras.')

caracteres = len(frase)

print(f'La frase tiene {caracteres} caracteres.')
```

## Ejercicio 9:

Escriba un programa que lea input del usuario: cantidad de días, horas, minutos y segundos y que en base a esos datos, calcule el total en segundos.

```python
dias = int(input('Dias: '))
horas = int(input('Horas: '))
minutos = int(input('Minutos: '))
segundos = int(input('Segundos: '))

dias_a_segundos = dias*24*60*60 # Convierto los dias solicitados al usuario a segundos (1 dia tiene 24 horas, cada hora tiene 60 minutos y cada minuto tiene 60 segundos),
                                # y guardo el resultado en la variable dias_a_segundos.

horas_a_segundos = horas*60*60 # Repito con horas.

minutos_a_segundos = minutos*60 # Repito con minutos.

total = dias_a_segundos + horas_a_segundos + minutos_a_segundos + segundos # Sumo todos los valores que ahora ESTÁN EN SEGUNDOS guardados en cada una de las variables.

print('Total en segundos: ',total)
```

# Ejercicio 10:

Haga un programa que pregunte por el precio de una mercaderia y el porcentaje de descuento de la misma y que calcule el importe del descuento y el precio a pagar.

```python
mercaderia = float(input('Precio: '))
porcentaje_descuento = int(input('Descuento: '))

importe_descuento = (mercaderia*porcentaje_descuento)/100 # Usando regla de 3, encuentro el descuento.

precio_pagar = mercaderia - importe_descuento # Guardo en precio_pagar el precio a pagar, que seria el descuento restado al precio de la mercaderia r

print('Importe descontado: ', importe_descuento)
print('Precio a pagar: ', precio_pagar)
```
