## Definición 

Este tipos de funciones son de Python o de Módulos de Python que podemos importar, pero también podemos hacer *nuestras propias funciones*. Para eso vamos usar la palabra clave `def`, seguida del nombre de la función y los argumentos que recibe.

Como por ejemplo, la siguiente función imprime el mensaje "¡Hola, mundo!" cuando se llama:

```python
def hola_mundo():
    print("¡Hola mundo!")
```

Para llamar a la función, simplemente escribe su nombre seguido de paréntesis:

```python
hola_mundo()
```

Como podemos ver la función es simple e imprime en pantalla lo esperado. Esta función no tiene ningún argumento. 

Las funciones también pueden tener argumentos, que son parámetros o valores que debemos pasan a la función cuando son llamas. Los argumentos se definen entre paréntesis después del nombre de la función. 

Por ejemplo, la siguiente función toma un parámetro llamado "nombre" y lo utiliza para imprimir un saludo personalizado:

```python
def saludar(nombre):
    print("¡Hola, " + nombre + "!")
```

Para llamar a esta función, se debe proporcionar un valor para el parámetro "nombre":

```python
saludar("Juan")
```

Además, las funciones en Python pueden devolver un valor usando la palabra clave "return". El valor devuelto puede ser utilizado por el programa que llama a la función. Por ejemplo, la siguiente función toma dos parámetros y devuelve su suma:

```python
def sumar(a, b):
    return a + b
```

Para utilizar el valor devuelto por la función, se debe asignar el resultado a una variable:

```python
resultado = sumar(2, 3)
print(resultado)
```

### Argumentos indeterminados

Una función puede usar parámetros indererminados, para eso, en Python, usamos la expresión `*args`. Por ejemplo:

```python
def sumar(*args):
  resultado = 0
  for valor in args:
      resultado = resultado + valor  
  return resultado

total = sumar(4, 2, 6, 9)
print(total)
```

```python
print(sumar(1,2))
```

```python
print(sumar(4,7,14,3,2,8,5,7))
```

Aquí la función identifica a los argumentos por la posición que ellos ocupan `args` (que viene de 'arguments' en ingles que significa argumentos o valores) es una tupla. 

Pero también podemos identificarlos por nombre (clave-valor o en inglés 'keyword args') usando el comando `kwargs`. Por ejemplo:

```python
def indeterminados_nombre(**kwargs):
    print(kwargs)
    print(type(kwargs))

indeterminados_nombre(n=5, c="Hola", l=[1,2,3,4,5])   

```

Aquí kwargs es un diccionario, y por ejemplo podriamos tener una función que itere sobre las llaves e imprima el valor de la siguiente forma:

```python
def func(**kwargs):
    for kwarg in kwargs:
        print(kwarg, "=>", kwargs[kwarg])

func(n=4, c='Hola mundo', a = [2,4,6])
```

### Retorno de múltiples valores 

Hasta ahora vimos funciones que hacian algo o que nos retornaban un único valor. Pero en Python podemos devolver múltiples valores. 

Veamos el siguiente ejemplo:

```python
def CuadradoCubo(n):
  cuadrado = n ** 2
  cubo = n ** 3
  return cuadrado, cubo

a = CuadradoCubo(5)
print(a,type(a))
```

### Función recursiva

Las funciones recursivas son aquellas que se llaman a sí mismas dentro de su propia definición. En Python, se pueden utilizar para resolver problemas que se pueden dividir en problemas más pequeños del mismo tipo.

Por ejemplo, supongamos que queremos calcular el factorial de un número entero. El factorial de un número es el producto de todos los números enteros desde 1 hasta ese número. 

El factorial de 3 lo escribimos así: 3! = 3 \* 2 \* 1 = 6

Entonces, 4! = 4 \* 3 \* 2 \* 1

Y 5! = 5 \* 4 \* 3 \* 2 \* 1 

Queremos una función que calcule el factorial de n, eso es: n! = n \* (n-1) \* (n-2) \* ...  \* 1 

*Nota: 0! = 1*


Podemos escribir una función recursiva para calcular el factorial de la siguiente manera:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(5)
```

El primer llamado a la función es `factorial(5)`. Como `5` no es igual a `0`, la función devuelve `5 * factorial(4)`. El siguiente llamado a la función es `factorial(4)`. De nuevo, como `4` no es igual a `0`, la función devuelve `4 * factorial(3)`. Este proceso continúa hasta que la función `factorial(1)` se llama, en cuyo caso devuelve `1`. Luego, las funciones se devuelven en orden inverso, de modo que `factorial(2)` devuelve `2 * 1 = 2`, `factorial(3)` devuelve `3 * 2 = 6`, `factorial(4)` devuelve `4 * 6 = 24`, y finalmente `factorial(5)` devuelve `5 * 24 = 120`.

* Ventajas de la recursividad:
> * Las funciones recursivas hacen que el código se vea limpio y elegante
> * Una tarea compleja se puede dividir en subproblemas más simples mediante la recursividad.
> * La generación de secuencias es más fácil con la recursivad que con algunas interaciones anidadas.

* Desventajas de la recursividad:
> * Algunas veces la lógica detrás de la recursividad es más dificil de seguir.
> * Las llamadas recursivas pueden ser ineficaces en lo que respecta a memoria y tiempo de ejecución.
> * Dependiendo de la complejidad, las funciones recursivas pueden ser mas difíciles de debuggear. 

### Funciones anidadas o funciones internas

La definición de una función dentro de otra función es útil en situaciones en las que una función necesita utilizar una función auxiliar que sólo se utiliza dentro de la función principal. En lugar de definir la función auxiliar fuera de la función principal y hacer que esté disponible en todo el programa, podemos definirla dentro de la función principal para limitar su alcance.

Por ejemplo queremos una función que calcule la suma de los cuadros de los números de una lista, sin usar NumPy. 

```python
def calcular_suma_cuadrados(lista):
    def cuadrado(numero):
        return numero ** 2

    suma = 0
    for num in lista:
        suma += cuadrado(num)
    return suma
```

```python
# Ejemplo de uso
lista = [1, 2, 3, 4, 5]
resultado = calcular_suma_cuadrados(lista)
print(resultado)
```

En este ejemplo, la función `calcular_suma_cuadrados` define una función anidada llamada `cuadrado`. Esta función interna toma un número y devuelve su cuadrado. La función principal `calcular_suma_cuadrados` utiliza la función `cuadrado` para calcular la suma de los cuadrados de los elementos de una lista. En este caso, la lista es `[1, 2, 3, 4, 5]`, por lo que la función `calcular_suma_cuadrados` devuelve la suma de los cuadrados de estos números, que es `1 + 4 + 9 + 16 + 25 = 55`.

Es importante tener en cuenta que la función interna `cuadrado` sólo es visible dentro de la función principal `calcular_suma_cuadrados`. No se puede acceder a ella desde fuera de la función. Esto significa que la función `cuadrado` no interfiere con otras partes del programa y sólo se utiliza cuando se llama a `calcular_suma_cuadrados`.

* La función termina al llegar a un `return`, devolviendo los valores correspondientes. Si hay código después de ello, ¡no se va a ejecutar!

```python
def sumar(a, b):
  s = a + b
  return s
  print('La suma es', s)

sumar(2,2)
```

> Esto es particularmente relevante en el caso de la función tenga varios `return`s:

```python
def buscar_pos(lista, a):
  '''Busca la posición de a en la lista. Si no está devulve None'''
  for i, e in enumerate(lista):
    if e == a:
      return i
  return

print(buscar_pos([1, 6, 3, 4, 3], 3)) # ¡Sólo devulve la primera posición!
print(buscar_pos([1, 6, 3, 4, 3], 7))
```

### Ejercicio 4 
(era el ejercicio opcional en la Semana 3)

Supongamos que tenemos una lista de números enteros, y queremos encontrar todas las subsecuencias de la lista que sumen un número objetivo dado. Para hacerlo, vamos a crear una función que tome como argumentos la lista de números y el número objetivo, y devuelva una lista con todas las subsecuencias que suman el número objetivo.

Ayuda: Por ejemplo, si le damos a tu programa la lista de números [1, 2, 3, 4, 5] y queremos encontrar todas las subsecuencias que suman 7 (número objetivo dado), la función debería devolver las siguientes listas [2, 5], [3, 4], [1, 2, 4].

```python
def subsecuencias_suma(lista, objetivo):
    def backtrack(inicio, sumaval, subsecuencia_actual):
        if sumaval == objetivo:
            res.append(subsecuencia_actual)
            return
        for i in range(inicio, len(lista)):
            if sumaval + lista[i] <= objetivo:
                backtrack(i+1, sumaval+lista[i], subsecuencia_actual+[lista[i]])

    res = []
    backtrack(0, 0, [])
    return res
```

```python
lista = [1, 2, 3, 4, 5]
objetivo = 7
resultados = subsecuencias_suma(lista, objetivo)
print(resultados)
```

### Atrapar y administrar excepciones

Las excepciones pueden ser atrapadas y administradas. Para atrapar una excepción, se usan los comandos `try` - `except`.

```python
numero_valido=False
while not numero_valido:
    try:
        a = input('Ingresá un número entero: ')
        n = int(a)
        numero_valido = True
    except ValueError:
        print('No es válido. Intentá de nuevo.')
print(f'Ingresaste {n}.')
```

Si en este ejemplo ingresas por ejemplo una letra, el comando `n = int(a)` genera una excepción de tipo ValueError: el comando `numero_valido = True` no se ejecuta, la excepción es atrapada por el except `ValueError` y el ciclo se repite. Probalo ingresando letras, números con decimales y números enteros.

Suele ser difícil saber exactamente qué tipo de errores pueden ocurrir por adelantado. Para bien o para mal, la administración de excepciones suele ir creciendo a medida que un programa va generando errores inesperados (al mejor estilo: "Uh! Me olvidé de que podía pasar esto. Deberíamos preverlo y administrarlo adecuadamente para la próxima").
