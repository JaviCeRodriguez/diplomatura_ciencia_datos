# Ejercicio 3

Escribir un programa que pregunte una fecha en formato dd/mm/aaaa y muestre por pantalla la misma fecha en formato dd de \<mes\> de aaaa donde \<mes \> es el nombre del mes. Para ello cree un diccionario que relacione el valor de mm con el nombre del mes. 

````python
# Creamos un diccionario que relacione los números de mes con sus nombres. Usamos de key las strings, porque despues cuando el prompteamos al usuario, input va a devolver todo en strings.
meses = {
    "01": "enero",
    "02": "febrero",
    "03": "marzo",
    "04": "abril",
    "05": "mayo",
    "06": "junio",
    "07": "julio",
    "08": "agosto",
    "09": "septiembre",
    "10": "octubre",
    "11": "noviembre",
    "12": "diciembre"
}

# Pedimos al usuario que ingrese una fecha en formato dd/mm/aaaa
fecha = input("Ingresa una fecha en formato dd/mm/aaaa: ")

# Separamos la fecha en día, mes y año utilizando el método split() de las cadenas
dia, mes, anio = fecha.split("/") # Notar que estamos usando 

# Mostramos la fecha en el formato solicitado utilizando los valores del diccionario y f-strings
print(f"{dia} de {meses[mes]} de {anio}") # Notar que acá usamos directamente el diccionario para llamar al nombre del mes.
````

# Ejercicio 4

Escribe un programa que pida al usuario que ingrese dos listas de números enteros separados por espacios. Luego, crea dos conjuntos con los elementos de cada lista y muestra por pantalla los siguientes conjuntos:

    1. La unión de los dos conjuntos
    2. La intersección de los dos conjuntos
    3. Los elementos que están en el primer conjunto pero no en el segundo
    4. Los elementos que están en el segundo conjunto pero no en el primero

````python
# Pedir al usuario que ingrese las dos listas de números enteros separados por espacios
list1 = input("Ingrese la primera lista de números enteros separados por espacios: ").split(' ')
list2 = input("Ingrese la segunda lista de números enteros separados por espacios: ").split(' ')

# Convertir las listas a conjuntos
set1 = set(list1)
set2 = set(list2)

# Mostrar la unión de los dos conjuntos
union_set = set1 | set2
print("La unión de los dos conjuntos es:", union_set)

# Mostrar la intersección de los dos conjuntos
intersection_set = set1 & set2
print("La intersección de los dos conjuntos es:", intersection_set)
# Mostrar los elementos que están en el primer conjunto pero no en el segundo
difference_set1 = set1 - set2
print("Los elementos que están en el primer conjunto pero no en el segundo son:", difference_set1)

# Mostrar los elementos que están en el segundo conjunto pero no en el primero
difference_set2 = set2 - set1
print("Los elementos que están en el segundo conjunto pero no en el primero son:", difference_set2)
````

# Ejercicio 5

Dadas las listas de nombres y edades:
nombre=[Juan, Ana, Marcela, Claudio, Nicolás, Cecilia, Tomas, Pedro]
edades=[25, 15, 13, 21, 45, 32, 17, 33]
1. Crear una lista de tuplas que contenga los nombres y las edades de las personas mayores de 18 años. Atención, usar la función `zip()`.
2. Ordenar la lista resultante de forma alfabética por el nombre de las personas.

````python
# Definir las listas de nombres y edades
nombres = ['Juan', 'Ana', 'Marcela', 'Claudio', 'Nicolás', 'Cecilia', 'Tomas', 'Pedro']
edades = [25, 15, 13, 21, 45, 32, 17, 33]

# Crear una lista vacía para almacenar las tuplas
mayores_de_edad = []

# Iterar a través de las listas de nombres y edades con la función zip
for nombre, edad in zip(nombres, edades):
    # Verificar si la persona es mayor de 18 años
    if edad > 18:
        # Agregar una tupla con el nombre y edad de la persona a la lista de mayores de edad
        mayores_de_edad.append((nombre, edad))

# Tambien lo podriamos haber hecho con comprensión de listas...
# mayores_de_edad = [(nombre,edad) for nombre,edad in zip(nombres,edades) if edad > 18]

# Ordenar la lista de tuplas en orden alfabético por el nombre de la persona
mayores_de_edad_ordenados = sorted(mayores_de_edad)
# Acá vemos que sorted lo que hace es ordenar la lista en base al primer elemento de cada tuplas, que en este caso es el nombre,
# y que justo cuadra con la implementación que queremos hacer. A priori esto no tendria porque ser asi, podria tener otro criterio
# o bien yo podria haber armado las tuplas al revés y no me funcionaria... 
# Moraleja: No asumas nada, probá... pero probá bien hasta que estés seguro que el programa hace lo que necesitas que haga siempre.
# Sino vas a dejar bugs potenciales, que a la larga podrian ser muy dificiles de detectar.

# Imprimir la lista de tuplas resultante
print(mayores_de_edad_ordenados)
````

# Ejercicio 7

Escribe un programa que lea una lista de palabras separadas por espacios y cree una tupla con ellas. Luego, muestra por pantalla la tupla creada y la cantidad de veces que aparece la letra más repetida en todas las palabras.




````python
# Pedir al usuario que ingrese una lista de palabras separadas por espacios
palabras = input("Ingrese una lista de palabras separadas por espacios: ")

# Dividir las palabras en una lista
lista_palabras = palabras.split()

# Convertir la lista en una tupla
tupla_palabras = tuple(lista_palabras)

# Inicializar un diccionario para almacenar la frecuencia de cada letra
frecuencia_letras = {}

# Iterar sobre cada palabra en la tupla
for palabra in tupla_palabras:
    # Iterar sobre cada letra en la palabra
    for letra in palabra:
        # Si no esta todavía la letra en el diccionario, la crea.
        if letra not in frecuencia_letras.keys():
            frecuencia_letras[letra] = 1
        else:
            # Incrementar la frecuencia de la letra en el diccionario
            frecuencia_letras[letra] += 1

# Encontrar la frecuencia más alta para despues poder todas las letras que llegan a esta frecuencia.
frecuencia_mas_alta = max(frecuencia_letras.values())

# Encontrar todas las letras con la frecuencia más alta
letras_mas_repetidas = []
for letra, frecuencia in frecuencia_letras.items():
    if frecuencia == frecuencia_mas_alta:
        letras_mas_repetidas.append(letra)

# Mostrar la tupla de palabras y las letras más repetidas con su frecuencia
print("Tupla de palabras:", tupla_palabras)

# Si la frecuencia mas alta es 1, entonces todas las letras se repiten 1 vez y no hay letra más repetida.
if frecuencia_mas_alta == 1: 
    print('Todas las letras se repiten por igual')
# Si solo hay una letra que reporta esa letra.
elif len(letras_mas_repetidas) == 1:
    print(f"La letra más repetida es '{letras_mas_repetidas[0]}' con una frecuencia de {frecuencia_mas_alta}.")
# Sino reporta todas las letras.
else:
    print(f"Las letras más repetidas son {letras_mas_repetidas} con una frecuencia de {frecuencia_mas_alta}.")
````

# Ejercicio 8

Imagina que eres el gerente de una tienda en línea y estás revisando tu base de datos de clientes. Descubres que algunos clientes se han registrado varias veces con la misma información de contacto, pero diferentes nombres de usuario. La lista con información de contacto que tenes es: 

```py
clientes = [

  (1, "Ana", "García", "ana@example.com", "Calle Falsa 123"), 

  (2, "Pedro", "Martínez", "pedro@example.com", "Avenida Siempreviva 742"),

  (3, "Luisa", "Fernández", "luisa@example.com", "Calle Falsa 123"),

  (4, "Juan", "Pérez", "juan@example.com", "Calle Falsa 123"), 

  (5, "Ana", "García", "ana@example.com", "Calle Falsa 123")
  
  ]
```
Donde el primer elemento de cada tupla es el numero de cliente, luego viene el nombre, apellido, email y dirección.

Para simplificar tu base de datos, decides eliminar los registros duplicados y ordenar la lista de clientes alfabéticamente. Escribe un código que haga esto.


````python
clientes = [

  (1, "Ana", "García", "ana@example.com", "Calle Falsa 123"), 

  (2, "Pedro", "Martínez", "pedro@example.com", "Avenida Siempreviva 742"),

  (3, "Luisa", "Fernández", "luisa@example.com", "Calle Falsa 123"),

  (4, "Juan", "Pérez", "juan@example.com", "Calle Falsa 123"), 

  (5, "Ana", "García", "ana@example.com", "Calle Falsa 123")

  ]

# Elimino los registros duplicados

# Genero una lista de tuplas sin el numero de clientes, para ir chequeando si el cliente esta duplicado. ¿Porque? Bueno, viendo clientes, me doy cuenta 
# lo que se repite es la informacion de contacto, pero el numero de cliente NO. Entonces tengo que encontrar una manera de poder comparar estas tuplas
# pero sin que se tenga en cuenta el numero de cliente.

clientes_sin_numero = []
for idx, cliente in enumerate(clientes, start=0): # Use enumerate para saber cual es el indice de la lista, y asi saber cual borrar si está duplicado.
    sin_numero = (cliente[1],cliente[2],cliente[3],cliente[4]) # Como no tiene el numero de cliente ahora si lo voy a poder usar para encontrar duplicados.
    if (cliente[1],cliente[2],cliente[3],cliente[4]) in clientes_sin_numero: # Si ya encontre el cliente en mi lista de clientes sin numeros, entonces ese cliente esta repetido!
        del clientes[idx] # Borro el cliente
    else: # Que pasa si no esta en la lista?... Lo agrego, asi al fin y al cabo voy a terminar con una lista (sin el numero de cliente) con todos los clientes sin repetir.
        clientes_sin_numero.append(sin_numero)

print(clientes) # Listo! Ya le saqué todos los duplicados.
````

El programa deberá guardar toda la información guardada, y además deberá indicar cuantas y cuales son las materias a las que hay por lo menos un estudiante inscrito.
