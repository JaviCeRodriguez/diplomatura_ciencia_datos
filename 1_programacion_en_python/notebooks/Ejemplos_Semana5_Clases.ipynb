### Otra forma
Otra forma de leer el archivo `estaciones_metereologicas.csv` puede ser creando un diccionario cuya primera columna sea la llave y el resto de las columnas el valor. 

```python
# creo un diccionatio vacio llamado data:
data = {}
# Abro el archivo con el path completo
with open('/content/drive/My Drive/Colab Notebooks/Data/estaciones_metereologicas.csv') as estaciones:
  filas = csv.reader(estaciones)
  encabezados = next(filas)
  for fila in filas:
    # uso la primer columna que tiene el nombre de la estación metereologica como llave
    key = fila[0]
    # el resto de las columnas seran una lista de valores del diccionario
    values = fila[1:]
    data[key] = [values]
print(data)
```

Dependiendo de la base de datos, no podriamos hacer esto. ¿Por qué?

Podriamos crear un dicicionario con una lista de diccionarios, por ejemplo:

```python
# creo un diccionario vacio llamado data
data = {}
# abro el archivo con el path completo
with open('/content/drive/My Drive/Colab Notebooks/Data/registro_temperatura365d_smn.csv') as f:
      filas = csv.reader(f)
      encabezados = next(filas)
      for fila in filas:
        key = fila[0]
        my_dict = {fila[3]: fila[1:3]}
        values = my_dict
        # verifico si la llave existe, si sí, hago un append
        if key in data:
            data[key].append(values)
        # sino, la creo
        else:
            data[key] = [values]
print(data)
```

Otra posibilidad es usar enumerate para no necesitar escribir los indices:

```python

# creo un diccionario vacio llamado data
data = {}
# abro el archivo con el path completo
with open('/content/drive/My Drive/Colab Notebooks/Data/registro_temperatura365d_smn.csv') as f:
    filas = csv.reader(f)
    encabezados = next(filas)
    for fila in filas:
        subdic = {}
        for i, valor in enumerate(fila):
            if i == 0:
                key = valor
            elif i == 3:
                subdic[valor] = fila[1:3]
        # verifico si la llave existe, si sí, hago un append
        if key in data:
            data[key].append(values)
        # sino, la creo
        else:
            data[key] = [values]

print(data)
```

En lugar de utilizar el condicional if key in data para verificar si la clave ya existe en el diccionario, se puede utilizar el método setdefault() para crear la clave en caso de que no exista.

```python
# creo un diccionario vacio llamado data
data = {}
# abro el archivo con el path completo
with open('/content/drive/My Drive/Colab Notebooks/Data/registro_temperatura365d_smn.csv') as f:
    filas = csv.reader(f)
    encabezados = next(filas)
    for fila in filas:
        subdic = {}
        for i, valor in enumerate(fila):
            if i == 0:
                key = valor
            elif i == 3:
                subdic[valor] = fila[1:3]
        # verifico que la llave existe y sino la creo        
        data.setdefault(key, []).append(subdic)

print(data)
```

Escribir un programa que pida un número entero entre 1 y 10 y guarde en un fichero con el nombre `tabla-n.txt` la tabla de multiplicar de ese número, done `n` es el número introducido`


```python
import os
# Conectar Drive a Colab
from google.colab import drive          # aqui le decimos al modulo google.colab que importe drive 
drive.mount("/content/drive/")   
# Cambio el directorio para Data
os.chdir('/content/drive/MyDrive/Colab Notebooks/Data/') # Uso absoluto para no tener problemas

n = int(input('Introduce un número entero entre 1 y 10: '))
nombre_fichero = 'tabla-' + str(n) + '.txt'
f = open(nombre_fichero, 'w')
for i in range(1, 11):
    f.write(str(n) + ' x ' + str(i) + ' = ' + str(n * i) + '\n')
f.close()
print('El archivo:',nombre_fichero,'fue creado en /content/drive/MyDrive/Colab Notebooks/Data/')
```
