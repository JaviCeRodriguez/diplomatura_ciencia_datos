## Complejidad de Algoritmos

La idea de esta clase es discutir y ejercitar el concepto de complejidad.
Hay libros enteros y cursos dedicados al concepto, pero brevemente pueden leer lo que dice Wikipedia al respecto, que está correcto: [Computational Complexity](https://en.wikipedia.org/wiki/Computational_complexity). 

Como vimos en la clase de presentación, hay algunos problemas que *a priori* parecen simples. Vamos a retomar en este notebook uno de los problemas presentados:

````python
# pregunta: existe algun subconjunto de números cuya suma sea 119?
numeros = [2, 3, 4, 5, 8, 9, 10, 19, 27, 76, 89]
suma = 119
````

Por ahora no vamos a resolver el problema completo solamente vamos a ocuparnos de los subconjuntos. 

Cuántos subconjuntos podemos tomar de la lista? Cuántas combinacionas hay tomando de a 2 elementos?
Tomando de a 3? De a 4?

Para calcular esto tenemos que entrar en el terreno de las combinatorias. En nuestro caso las combinaciones representan el número de maneras en las que se pueden seleccionar **r** objetos de un total de **n** objetos disponibles en los cuales el orden de selección no es de importancia. Esto se llama "Combinaciones sin repetición". Estos cálculos usan factoriales.  


image.png


En Python los podemos calcular aprovechando el modulo `math` usando `math.factorial()`

```
# Formula: n! / ( r! * (n-r)! )
# usando math.factorial() seria asi
import math
math.factorial(n) / ( math.factorial(r) * math.factorial(n-r) )
```

````python
import math
n = 11   # tenemos esta cantidad de numeros (el conjunto)
r = 2    # tomamos subconjuntos de este tamaño
combinatorias = math.factorial(n) / (math.factorial(r) * math.factorial(n-r))
print(combinatorias)
````

A continuación, las calculamos todas.

````python
import math

# el conjunto 
numeros = [2, 3, 4, 5, 8, 9, 10, 19, 27, 76, 89]
n = len(numeros)

# inicializo variables para guardar resultados
combinatorias = {}
suma_de_combinatorias = 0

def combinatoria(n,r):
  return math.factorial(n) / ( math.factorial(r) * math.factorial(n-r) )

# vamos a calcular todos los posibles tamaños de subconjuntos
for r in range(1,n+1):
  combinatorias[r] = combinatoria(n,r)
  suma_de_combinatorias += combinatorias[r]
  print("Combinatorias de a", r, "=", combinatorias[r])
print("Total de combinatorias:", suma_de_combinatorias)
````

## Alternativa

Para los que quieran saber más, esto se puede hacer más facil usando el módulo `itertools`.

itertools = Functions creating iterators for efficient looping (Ver *Combinatoric iterators* en [la documentación Python sobre itertools](https://docs.python.org/3/library/itertools.html).)

````python
# un desvio en el camino ...

import itertools

print("=====> Combinaciones de a 2")
for S in itertools.combinations(numeros,2):
  print(S)

print("=====> Combinaciones de a 3")
for T in itertools.combinations(numeros,3):
  print(T)
````

````python
# volviendo a nuestro problema 
numeros = [2, 3, 4, 5, 8, 9, 10, 19, 27, 76, 89]

# veamos los subconjuntos usando itertools
subconjuntos = 0

n = len(numeros)
for r in range(1,n+1):
  for subconjunto in itertools.combinations(numeros,r):
    print(subconjunto)
    subconjuntos += 1

print("Total subconjuntos:", subconjuntos, "| Total de numeros:", n)
````

## Volviendo a nuestro problema

Pregunta: existe algun subconjunto de números cuya suma sea 119? Retomando, lo que nos falta ahora es evaluar la suma para cada subconjunto. 

````python
# retomando ... 

# funcion para evaluar una suma
# recibe dos parametros (inputs):
#   un 'subconjunto' que es la lista de numeros que queremos evaluar
#   un 'resultado_deseado' que es lo que queremos que de la suma
def evaluar_suma(subconjunto, resultado_deseado):
  cuenta = 0     # inicializamos la cuenta en cero 
  estado = False # y el estado en FALSO. Lo vamos a cambiar a VERDADERO si la suma da lo que queremos 
  n = len(subconjunto)
  for i in range(0,n):
    cuenta = cuenta + subconjunto[i]
  if (cuenta == resultado_deseado):
    estado = True
  return estado

# funcion para iterar por todas las combinaciones de numeros
def iterar_combinaciones(conjunto, resultado_deseado):
  terminar = False
  iteraciones = 0
  n = len(conjunto)
  for r in range(1,n+1):
    for subconjunto in itertools.combinations(conjunto,r):
      iteraciones += 1   # para seguir el rastro de lo que hacemos
      resultado = evaluar_suma(subconjunto, resultado_deseado)
      if resultado == True:
        terminar = True
        print("Subconjunto:", subconjunto, "| Suma", suma, "?", resultado, "| Iteraciones:", iteraciones)
        break 
    if terminar:
      break

conjunto = [2, 3, 4, 5, 8, 9, 10, 19, 27, 76, 89]
resultado_deseado = 119

# a calcular!
iterar_combinaciones(conjunto, resultado_deseado)
````

### Mejor caso y Peor Caso

Dado un algoritmo cualquiera, el consumo de recursos (ej tiempo que tarda en correr, uso de CPU, RAM, almacenamiento, etc) puede evaluarse para el mejor caso, peor caso y algun caso promedio (en algun lado entre el mejor y el peor). 

Veamoslo con ejemplos:

````python
# mejor caso
numeros = [3, 27, 89, 2, 4, 5, 8, 9, 10, 19, 76]
suma = 3
iterar_combinaciones(numeros, suma)
````

````python
# peor caso
numeros = [2, 4, 5, 8, 9, 10, 19, 76, 3, 27, 89]
suma = 252
iterar_combinaciones(numeros, suma)
````

## Por qué este es el peor caso? Es siempre este el peor caso?

Así como está resuelto el problema, nuestro algoritmo va en orden creciente, evaluando primero los subconjuntos de 1 elemento (r = 1), después los de 2, etc. Como en este caso la suma es la de todos los numeros del conjunto inicial, **siempre** vamos a llegar al final (va a dar VERDADERO en la última combinación). 

Se puede hacer algo al respecto? Si! Randomizar. Pregunta: randomizar resuelve el problema del peor caso en forma generica (para cualquier conjunto y suma)? 

O solamente para esta instancia de este conjunto de numeros? (`[2, 4, 5, 8, 9, 10, 19, 76, 3, 27, 89]`), y esta suma (`252`)?

Les queda este código de ejemplo mostrativo: 

````python
# solucion alternativa randomizando el orden de evaluacion de subconjuntos
import random

# funcion para evaluar una suma
# recibe dos parametros (inputs):
#   un 'subconjunto' que es la lista de numeros que queremos evaluar
#   un 'resultado_deseado' que es lo que queremos que de la suma
def evaluar(subconjunto, resultado_deseado):
  cuenta = 0     # inicializamos la cuenta en cero 
  estado = False # y el estado en FALSO. Lo vamos a cambiar a VERDADERO si la suma da lo que queremos 
  n = len(subconjunto)
  for i in range(0,n):
    cuenta = cuenta + subconjunto[i]
  if (cuenta == resultado_deseado):
    estado = True
  return estado

# funcion para iterar por todas las combinaciones de numeros
def iterar_random(conjunto, resultado_deseado):
  terminar = False
  iteraciones = 0
  n = len(conjunto)
  valores_de_r = list(range(1,n+1))
  random.shuffle(valores_de_r)
  for r in valores_de_r:
    for subconjunto in itertools.combinations(conjunto,r):
      iteraciones += 1   # para seguir el rastro de lo que hacemos
      resultado = evaluar(subconjunto, resultado_deseado)
      if resultado == True:
        terminar = True
        print("Subconjunto:", subconjunto, "| Suma", suma, "?", resultado, "| Iteraciones:", iteraciones)
        break 
    if terminar:
      break

conjunto = [2, 3, 4, 5, 8, 9, 10, 19, 27, 76, 89]
resultado_deseado = 252

# a calcular! (Ojo que la randomizacion es diferente cada vez!!!)
iterar_random(conjunto, resultado_deseado)
````

## Costo computacional

El costo computacional de un algoritmo se mide en *recursos* que consume. Uno que ya vimos, es el **tiempo**. Pero hay otros costos que pueden medirse y evaluarse. No los vamos a ejercitar en esta clase pero sepan que un algoritmo puede ser rápido pero consumir mucha RAM, o mucho CPU, o GPU (cómputo en tarjetas gráficas), o mucho almacenamiento en disco. 

Dependiendo de las necesidades de cada algoritmo, si el programa tarde mucho en ejecutarse y dar el resultado, es posible que esté consumiendo o siendo limitado por algunos de estos recursos. Ergo, el tiempo es una manera *indirecta* de medir el consumo de recursos. ## Complejidad Constante O(c)

Un algoritmo tiene complejidad constante cuando es *constante* en el número de pasos y tiempo/recursos que consume, **sin importar** el tamaño del input. 

````python
# algoritmo de complejidad constante 

def algoritmo_constante(items):
    resultado = items[0] * items[0]
    return resultado

# probemoslo 
#%timeit algoritmo_constante([4, 5, 6, 8])
#%timeit algoritmo_constante([4, 5, 6, 8, 9, 10])
#%timeit algoritmo_constante([4, 5, 6, 8, 8, 9, 10, 12, 14, 16])
#%timeit algoritmo_constante([4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30])
````

````python
# tambien podemos correr %timeit con parametros y guardar el resultado 

input1 = [4, 5, 6, 8]
input2 = [4, 5, 6, 8, 9, 10]
input3 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16]
input4 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30]

# -n = nro de ejecuciones
# -r = nro de repeticiones
# -o = generar objeto como resultado
t1 = %timeit -n10000000 -r5 -o algoritmo_constante(input1)
t2 = %timeit -n10000000 -r5 -o algoritmo_constante(input2)
t3 = %timeit -n10000000 -r5 -o algoritmo_constante(input3)
t4 = %timeit -n10000000 -r5 -o algoritmo_constante(input4)

# el objeto tiene metodos y atributos, ej
# t1.best
# t1.worst
# etc.
````

````python
# esto nos permite por ejemplo graficar 
# para visualizar lo que pasó 

import matplotlib.pyplot as plt
import numpy as np 

x = [len(input1), len(input2), len(input3), len(input4)]
y = [t1.best, t2.best, t3.best, t4.best]

# visualizar
fig, ax = plt.subplots()
ax.plot(x, y, 'b')
ax.set_xlabel('Input')
ax.set_ylabel('Segundos')
ax.set_ylim(0,np.max(y)*2)
ax.set_title('Complejidad Constante O(c)' )
plt.show()
````

## Complejidad Lineal O(n)

Este es el caso de algoritmos en donde los pasos ejecutados o recursos consumidos escalan linealmente con el input. 

````python
# algoritmo de complejidad lineal 

def algoritmo_lineal(items):
    for item in items:
        item + 1

# probemoslo
input1 = [4, 5, 6, 8]
input2 = [4, 5, 6, 8, 9, 10]
input3 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16]
input4 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30]

t1 = %timeit -n1000 -r5 -o algoritmo_lineal(input1)
t2 = %timeit -n1000 -r5 -o algoritmo_lineal(input2)
t3 = %timeit -n1000 -r5 -o algoritmo_lineal(input3)
t4 = %timeit -n1000 -r5 -o algoritmo_lineal(input4)

import matplotlib.pyplot as plt
import numpy as np 

x = [len(input1), len(input2), len(input3), len(input4)]
y = [t1.best, t2.best, t3.best, t4.best]
plt.plot(x, y, 'b')
plt.xlabel('Input')
plt.ylabel('Segundos')
plt.ylim(0,np.max(y)*1.2)
plt.title('Complejidad Lineal O(n)' )
plt.show()
````

## Complejidad Cuadrática O(n^2)



````python
def algoritmo_cuadratico(items):
  for item in items:
    for item2 in items:
      item + item2

# probemoslo
input1 = [4, 5, 6, 8]
input2 = [4, 5, 6, 8, 9, 10]
input3 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16]
input4 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30]
input5 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30, 34, 36, 39, 42]
input6 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30, 34, 36, 39, 42, 53, 57, 59, 65, 69]
input7 = [4, 5, 6, 8, 8, 9, 10, 12, 14, 16, 22, 26, 30, 34, 36, 39, 42, 53, 57, 59, 65, 69, 72, 75, 77, 79, 86, 92, 112]

t1 = %timeit -n100 -r5 -o algoritmo_cuadratico(input1)
t2 = %timeit -n100 -r5 -o algoritmo_cuadratico(input2)
t3 = %timeit -n100 -r5 -o algoritmo_cuadratico(input3)
t4 = %timeit -n100 -r5 -o algoritmo_cuadratico(input4)
t5 = %timeit -n100 -r5 -o algoritmo_cuadratico(input5)
t6 = %timeit -n100 -r5 -o algoritmo_cuadratico(input6)
t7 = %timeit -n100 -r5 -o algoritmo_cuadratico(input7)

import matplotlib.pyplot as plt
import numpy as np 

x = [len(input1), len(input2), len(input3), len(input4), len(input5), len(input6), len(input7)]
y = [t1.best, t2.best, t3.best, t4.best, t5.best, t6.best, t7.best]
plt.plot(x, y, 'b')
plt.xlabel('Input')
plt.ylabel('Segundos')
plt.ylim(0,np.max(y)*1.2)
plt.title('Complejidad Cuadrática O(n^2)' )
plt.show()
````

## Análisis de Complejidad de Algoritmos

Cuando tenemos un algoritmo más complejo, con varias funciones y métodos, el análisis de la complejidad debe hacerse **por partes**.

Tomemos este ejemplo:

````python
# cual es la complejidad de este algoritmo?

def algoritmo_complejo(items):

    for i in range(5):
        print("Python es lo más")

    for item in items:
        print(item)

    for item in items:
        print(item)

    print("O")
    print("O")
    print("O")

algoritmo_complejo([4, 5, 6, 8])
````

Al inicio de la función tenemos `for i in range(5)` que es de complejidad constante O(5) dado que se ejecutan 5 pasos sin importar el tamaño del input. 

A continuación tenemos `for item in items` que es de complejidad lineal O(n). Como tenemos dos lineas similares, tenemos dos veces O(n). 

Finalmente tenemos `print("O")` que nuevamente es de complejidad constante. Como tenemos tres líneas similares, la complejidad es O(3). 

Sumando, tenemos: 
```
O(5) + O(n) + O(n) + O(3) = O(8) + O(2n)
```
